```python

import matplotlib.pyplot as plt

# Заданные значения
age = 30
retirement_age = 60
target_capital = 350000
annual_return_rate = 0.08

# Расчет количества лет до пенсии
years_to_retirement = retirement_age - age

# Расчет суммы инвестиций каждый год для достижения цели
investment_per_year = target_capital * annual_return_rate / ((1 + annual_return_rate) ** years_to_retirement - 1)

# Создание списка лет
years = list(range(age, retirement_age + 1))

# Создание списка значений капитала
capital = []
current_capital = 0

for year in years:
    current_capital = (current_capital + investment_per_year) * (1 + annual_return_rate)
    capital.append(current_capital)

# Построение диаграммы
plt.plot(years, capital)
plt.xlabel('Возраст')
plt.ylabel('Капитал')
plt.title('Достижение целевого капитала к пенсии')
plt.grid(True)
plt.show()

print("Сумма инвестиций каждый год:", investment_per_year)

```
